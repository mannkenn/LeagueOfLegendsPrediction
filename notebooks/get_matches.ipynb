```python
# Import goodies
import pandas as pd
import numpy as np
import requests
import time
```

```python
data = pd.read_csv(r"C:\Users\toirt\LeagueOfLegendsPrediction\data\puuids.csv")
df = pd.DataFrame(data)
puuids_list = df['puuids'].tolist()
df.head()
```

```python
api_key = 'RGAPI-284d0eeb-b757-4ea1-ad73-bc8c5ac886c2'
api_url = 'https://americas.api.riotgames.com/lol/match/v5/matches/by-puuid'
```

```python
def get_match_id(api_key, api_url, puuid, count):
    '''
    Get match id from player puuid 
    Count -> number of match ids to return
    '''
    match_url = '%s/%s/ids?type=ranked&start=0&count=%d&api_key=%s' %(api_url, puuid, count, api_key)
    while True:
        resp = requests.get(match_url)
        
        if resp.status_code == 429:
            print('Sleepn')
            time.sleep(10)
            continue
    
        data = resp.json()
        return data
```

```python
match_ids = []
for puuid in puuids_list:
    match_id = get_match_id(api_key, api_url, puuid, 100)
    match_ids.extend(match_id)
    
# Export match_ids as csv
df = pd.DataFrame(match_ids, columns = ['match_ids'])
df.to_csv(r"C:\Users\toirt\LeagueOfLegendsPrediction\data\match_ids.csv", index = False)
```
